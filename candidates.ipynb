```python
#! /usr/bin/python3
import random
import time
import datetime
import pandas as pd
from enum import Enum



class Student:
    def __init__(self, firstName, lastName, testId, adress, math, english, socialStudies):
        self.firstName = firstName
        self.lastName = lastName
        self.testId = testId
        self.adress = adress
        self.math = math
        self.english = english
        self.socialStudies = socialStudies

    def report_score(self):
        #print(self.firstName, self.lastName, "Math: " + str(self.math) + " English: " + str(self.english) + " Social Studies: " + str(self.socialStudies))
        print('{:10s} {:10s} Math: {:3d}  English: {:3d}  Social Studies: {:3d}'.format(self.firstName, self.lastName, self.math ,self.english, self.socialStudies))
        
    def as_dict(self):
        return {'firstname': self.firstName, 
                'lastname': self.lastName, 
                'testId': self.testId,
                'address': self.adress,
                'math': self.math,
                'english': self.english,
                'socialstudies': self.socialStudies}


def generate_score_report(students):
    mathList = []
    englishList = []
    socialStudiesList = []
    mathsum = 0
    englishsum = 0
    socstudiessum = 0
    for person in students:
        mathList.append(person.math)
        mathsum += person.math
        englishList.append(person.english)
        englishsum += person.english
        socialStudiesList.append(person.socialStudies)
        socstudiessum += person.socialStudies
        
    #mean calc
    mathMean = mathsum/len(mathList)
    englishMean = englishsum/len(englishList)
    socialStudiesMean = socstudiessum/len(socialStudiesList)

    print("math min: " + str(min(mathList)) + "\n" + "math max: " + str(max(mathList)) + "\n" + "math mean: " + str(mathMean))
    print("english min: " + str(min(englishList)) + "\n" + "english max: " + str(max(englishList)) + "\n" + "english mean: " + str(englishMean))
    print("socialStudies min: " + str(min(socialStudiesList)) + "\n" + "socialStudies max: " + str(max(socialStudiesList)) + "\n" + "socialStudies mean: " + str(socialStudiesMean))





```

```python
if __name__ == "__main__":

    # reading this file to get a list of names. 
    # https://gist.githubusercontent.com/namuol/2657233/raw/74135b2637e624848c163759be9cd14ae33f5153/presidents.csv

    randomTestId = 0
    candidates = []

    fd = open('presidents.csv', 'r')
    for line in fd:
        # ignore first line, because it is a header
        if 'Presidency' in line:
            continue
        else:
            words = line.split(',')
            name = words[1]
            nm =  name.split()
            firstname = nm[0]
            lastname = nm[1]
            testid = randomTestId
            address = "1600 Pennsylvania Ave., NW. Washington , DC 20500"
            math = random.randint(10,100)
            english = random.randint(10,100)
            socialstudies = random.randint(10,100)
            randomTestId += 1
            candidate = Student(firstname, lastname, testid, address, math, english, socialstudies)
            candidate.report_score()
        candidates.append(candidate)

    generate_score_report(candidates)
    
    
```

```python
df = pd.DataFrame([x.as_dict() for x in candidates])
```
